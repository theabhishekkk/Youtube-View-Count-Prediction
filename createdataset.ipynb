```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Generating date range for 2024
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date)

# Generating random view counts for each day
np.random.seed(42)  # For reproducibility
view_counts = np.random.randint(100, 5000, size=len(date_range))

# Creating  DataFrame
data = pd.DataFrame({'Date': date_range,'ViewCount': view_counts})
```
