```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Generating date range for 2024
start_date = datetime(2024, 1, 1)
end_date = datetime(2024, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date)

# Generating random view counts for each day
np.random.seed(42)  # For reproducibility
view_counts = np.random.randint(1000, 5000, size=len(date_range))
likes = np.random.randint(500, 1000, size=len(date_range))
comments = np.random.randint(10, 50, size=len(date_range))

# Creating a DataFrame
data = pd.DataFrame({
    'Date': date_range,
    'ViewCount': view_counts,
    'Likes': likes,
    'Comments': comments
})
```

```python
#change data frame to csv file and then downloading it from folder
data.to_csv('data.csv', index=False)
```
